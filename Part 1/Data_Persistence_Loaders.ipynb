```python
import pandas as pd
```

```python
# Load the datasets
calendar_df = pd.read_csv('/content/drive/MyDrive/Hotel_Crawler/calendar_small.csv')
data_scraping_df = pd.read_csv('/content/drive/MyDrive/Hotel_Crawler/data_scraping.csv', encoding='ISO-8859-1')

# Display the first few rows of each dataframe
calendar_df_head = calendar_df.head()
data_scraping_df_head = data_scraping_df.head()

(calendar_df_head, data_scraping_df_head)
```

```python
# Extract check-in date from the 'url' column
data_scraping_df['date'] = data_scraping_df['url'].str.extract(r'check_in=(\d{4}-\d{2}-\d{2})')

# Display the first few rows to verify the new 'date' column
data_scraping_df_head_with_date = data_scraping_df.head()
data_scraping_df_head_with_date
```

```python
# Extract listing ID from the 'url' column
data_scraping_df['listing_id'] = data_scraping_df['url'].str.extract(r'/rooms/(\d+)?')

# Display the first few rows to verify the new 'listing_id' column
data_scraping_df_head_with_listing_id = data_scraping_df.head()
data_scraping_df_head_with_listing_id
```

```python
# Check calendar_small dataframe columns to align with the task
calendar_columns_check = calendar_df.columns

# Prepare data_scraping dataframe with specified columns and adding 'available' column with 't'
data_scraping_prepared_df = data_scraping_df[['listing_id', 'date', 'price_alt']].copy()
data_scraping_prepared_df['available'] = 't'
data_scraping_prepared_df = data_scraping_prepared_df.rename(columns={'price_alt': 'price'})

# Since the actual merge strategy depends on the structure of calendar_small.csv, which we'll adjust according to the column names
calendar_columns_check
```

```python
# Prepare calendar_small dataframe for merge
calendar_prepared_df = calendar_df[['listing_id', 'date', 'available', 'price']]

# Since data_scraping_prepared_df already has the required format and 'available' set to 't', we can concatenate
# Note: Here we are concatenating instead of merging because it seems we are combining records rather than matching on keys
merged_df = pd.concat([calendar_prepared_df, data_scraping_prepared_df], ignore_index=True)

# Display the first few rows of the merged dataframe to verify
merged_df_head = merged_df.head()
merged_df_head
```

```python
merged_df.to_csv("Persistence_Loaded.csv", index=False)
```
